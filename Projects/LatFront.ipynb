```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
pledge = pd.read_csv("/Users/gracechi/Desktop/latfront.csv")

pledge['Theme'].value_counts()[:5].plot(kind = 'barh')

#X-Axis Label
plt.xlabel('Pledge Theme', fontsize = 16)
#Y-Axis Label
plt.ylabel('Frequency of Theme', fontsize = 16)

#Removes Spine
plt.spines['right'].set_visible(False)
plt.spines['top'].set_visible(False)
#Graph Title
plt.suptitle('Leaders at the Front Most Common Themes', fontsize = 20)
#Graph Dimensions
plt.gcf().set_size_inches(13,8)


#Shows graph
plt.show()

```
