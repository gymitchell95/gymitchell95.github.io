```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df_pledge = pd.read_csv("/Users/gracechi/Desktop/latfront.csv")

df_pledge['Theme'].value_counts()[:5].plot(kind = 'barh')

#X-Axis Label
plt.xlabel('Frequency of Theme', fontsize = 16)
#Y-Axis Label
plt.ylabel('Type of Pledge', fontsize = 16)


#Graph Title
plt.suptitle('Leaders at the Front Most Common Themes', fontsize = 20)
#Graph Dimensions
plt.gcf().set_size_inches(13,8)


#Shows graph
plt.show()

```
